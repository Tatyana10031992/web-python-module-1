# ЗАДАНИЕ: Консольное приложение для учета складских заявок на выдачу материалов

Необходимо разработать консольное приложение WarehouseRequest Console для учета материалов на складе и обработки заявок на их выдачу сотрудникам организации.

В компании есть внутренний склад, на котором хранятся различные материалы: бумага, картриджи, кабели, канцелярские товары, расходники, комплектующие и другие позиции. Сотрудники разных отделов могут создавать заявки на получение материалов. Кладовщик должен видеть остатки на складе, обрабатывать заявки, пополнять материалы и списывать их после выдачи.

Приложение должно позволять вести учет материалов, создавать заявки на выдачу, контролировать остатки, изменять статусы заявок и получать краткую статистику.

## Цель работы:
Создать консольную программу, которая позволяет добавлять материалы на склад, просматривать остатки, создавать и обрабатывать заявки на выдачу материалов, а также сохранять данные между запусками программы.

## Функциональные требования:

### 1. Главное меню

При запуске программы пользователь должен видеть меню:

```txt
=== WarehouseRequest Console ===

1. Добавить материал на склад
2. Показать все материалы
3. Создать заявку на выдачу
4. Показать все заявки
5. Найти заявку
6. Изменить статус заявки
7. Пополнить остаток материала
8. Списать материал по заявке
9. Показать статистику
10. Сохранить данные
0. Выход
```

Пользователь выбирает действие путем ввода номера пункта меню.

### 2. Добавление материала на склад

Программа должна позволять добавить новый материал на склад.

При добавлении материала пользователь вводит следующие данные:

```txt
- название материала
- категория материала
- единица измерения
- количество на складе
- минимальный остаток
```

**Категория материала может быть любой, например:**

- канцелярия
- расходники
- техника
- комплектующие
- упаковка
- другое

**Единица измерения может быть любой, например:**

- шт.
- пачка
- коробка
- метр
- кг

**Каждому материалу должен автоматически присваиваться уникальный номер.**

**Пример материала:**

```txt
Номер материала: 1
Название: Бумага A4
Категория: канцелярия
Единица измерения: пачка
Количество на складе: 25
Минимальный остаток: 5
```

### 3. Просмотр всех материалов

Программа должна выводить список всех материалов, которые есть на складе.

Для каждого материала необходимо отображать:

```txt
- номер материала
- название материала
- категория
- единица измерения
- количество на складе
- минимальный остаток
```

Если материалов нет, программа должна вывести сообщение:

`Материалы отсутствуют.`

Если количество материала меньше или равно минимальному остатку, рядом с материалом желательно выводить предупреждение:

`Требуется пополнение`

**Пример вывода:**

```txt
Номер материала: 2
Название: Картридж HP 106A
Категория: расходники
Единица измерения: шт.
Количество на складе: 2
Минимальный остаток: 3
Статус остатка: Требуется пополнение
```

### 4. Создание заявки на выдачу

Программа должна позволять создать заявку на выдачу материала сотруднику.

При создании заявки пользователь вводит:

```txt
- ФИО сотрудника
- отдел сотрудника
- номер материала
- количество материала
- причина выдачи
- статус заявки
```

**Статус заявки может принимать одно из значений:**

- новая
- одобрена
- отклонена
- выполнена

**Каждой заявке должен автоматически присваиваться уникальный номер.**

Перед созданием заявки программа должна проверить:

1. Существует ли материал с указанным номером.
2. Количество запрашиваемого материала должно быть больше 0.
3. Запрашиваемое количество не должно превышать количество материала на складе.

Если все проверки пройдены, заявка создается.

**Пример заявки:**

```txt
Номер заявки: 1
Сотрудник: Иванов Иван Иванович
Отдел: Бухгалтерия
Материал: Бумага A4
Количество: 3 пачка
Причина выдачи: Печать отчетных документов
Статус: новая
```

### 5. Просмотр всех заявок

Программа должна выводить список всех созданных заявок.

Для каждой заявки необходимо отображать:

```txt
- номер заявки
- ФИО сотрудника
- отдел
- название материала
- количество материала
- причина выдачи
- статус заявки
```

Если заявок нет, программа должна вывести сообщение:

`Заявки отсутствуют.`

### 6. Поиск заявки

Программа должна позволять искать заявки по разным параметрам.

Необходимо реализовать поиск:

```txt
- по ФИО сотрудника
- по отделу
- по названию материала
- по статусу заявки
```

При выборе пункта `Найти заявку` пользователь должен выбрать критерий поиска, затем ввести значение для поиска.

Пример меню поиска:

```txt
1. Поиск по сотруднику
2. Поиск по отделу
3. Поиск по материалу
4. Поиск по статусу
```

Если подходящие заявки найдены, программа выводит их список.

Если заявки не найдены, программа выводит сообщение:

`Заявки не найдены.`

### 7. Изменение статуса заявки

Программа должна позволять изменить статус существующей заявки.

Для изменения статуса пользователь вводит номер заявки, после чего выбирает новый статус:

```txt
1. новая
2. одобрена
3. отклонена
4. выполнена
```

Если заявка с указанным номером существует, программа изменяет ее статус и выводит сообщение:

`Статус заявки успешно изменен.`

Если заявка не найдена, программа выводит сообщение:

`Заявка с указанным номером не найдена.`

### 8. Пополнение остатка материала

Программа должна позволять увеличить количество материала на складе.

Для пополнения остатка пользователь вводит:

```txt
- номер материала
- количество для добавления
```

Если материал с указанным номером существует, программа увеличивает его остаток и выводит сообщение:

`Остаток материала успешно пополнен.`

Если материал не найден, программа выводит сообщение:

`Материал с указанным номером не найден.`

Если пользователь вводит отрицательное количество или ноль, программа должна вывести сообщение:

`Количество для пополнения должно быть больше нуля.`

### 9. Списание материала по заявке

Программа должна позволять списать материал со склада по заявке.

Для списания пользователь вводит номер заявки.

Списание возможно только если:

```txt
- заявка существует
- статус заявки: одобрена
- на складе достаточно материала
```

После успешного списания:

```txt
- количество материала на складе уменьшается
- статус заявки меняется на выполнена
```

Если заявка не найдена, программа выводит сообщение:

`Заявка с указанным номером не найдена.`

Если заявка не одобрена, программа выводит сообщение:

`Списание возможно только по одобренной заявке.`

Если материала на складе недостаточно, программа выводит сообщение:

`Недостаточно материала на складе.`

Если списание выполнено успешно, программа выводит сообщение:

`Материал успешно списан по заявке.`

### 10. Статистика

Программа должна выводить краткую статистику:

```txt
- общее количество материалов
- общее количество заявок
- количество новых заявок
- количество одобренных заявок
- количество отклоненных заявок
- количество выполненных заявок
- количество материалов, требующих пополнения
- общее количество единиц материалов на складе
```

Пример вывода:

```txt
=== Статистика ===

Всего материалов: 8
Всего заявок: 15
Новые заявки: 4
Одобренные заявки: 3
Отклоненные заявки: 2
Выполненные заявки: 6
Материалы, требующие пополнения: 2
Всего единиц материалов на складе: 147
```

### 11. Сохранение данных

Программа должна сохранять данные о материалах и заявках в файл.

Формат файла можно выбрать самостоятельно:

- txt
- csv
- json

После сохранения программа должна вывести сообщение:

`Данные успешно сохранены.`

### 12. Загрузка данных

При запуске программы необходимо загружать ранее сохраненные материалы и заявки из файла, если такой файл существует.

### 13. Выход из программы

При выборе пункта `0. Выход` программа должна завершить работу.

Перед выходом желательно выполнить автоматическое сохранение данных в файл.
