# ЗАДАНИЕ: Консольное приложение для учета заявок на обслуживание оборудования

Необходимо разработать консольное приложение ServiceDesk Console для учета заявок на техническое обслуживание оборудования в организации.

В организации сотрудники используют различное оборудование: ноутбуки, компьютеры, принтеры, роутеры, сканеры и другую технику. Если оборудование выходит из строя или требует обслуживания, сотрудник оставляет заявку. Приложение должно позволять вести учет таких заявок, отслеживать их статус и получать краткую статистику.

## Цель работы:
Создать консольную программу, которая позволяет добавлять, просматривать, искать, редактировать и удалять заявки на обслуживание оборудования, а также сохранять данные между запусками программы.

## Функциональные требования:

### 1. Главное меню

При запуске программы пользователь должен видеть меню:

```txt
=== ServiceDesk Console ===

1. Добавить заявку
2. Показать все заявки
3. Найти заявку
4. Изменить статус заявки
5. Удалить заявку
6. Показать статистику
7. Сохранить данные
0. Выход
```
Пользователь выбирает действие путем ввода номера пункта меню.

### 2. Добавление заявки

Программа должна позволять добавить новую заявку.

При создании заявки пользователь вводит следующие данные:

```txt
- ФИО сотрудника
- отдел сотрудника
- тип оборудования
- инвентарный номер оборудования
- описание проблемы
- приоритет заявки
- статус заявки
```

**Приоритет заявки может принимать одно из значений:**

- низкий
- средний
- высокий

**Статус заявки может принимать одно из значений:**

- новая
- в работе
- выполнена

**Каждой заявке должен автоматически присваиваться уникальный номер.**

**Пример заявки:**
```txt
Номер заявки: 1
Сотрудник: Иванов Иван Иванович
Отдел: Бухгалтерия
Оборудование: Принтер
Инвентарный номер: PR-1023
Описание проблемы: Не печатает документы
Приоритет: высокий
Статус: новая
```

### 3. Просмотр всех заявок

Программа должна выводить список всех созданных заявок.

Для каждой заявки необходимо отображать:
```txt
- номер заявки
- ФИО сотрудника
- отдел
- тип оборудования
- инвентарный номер
- описание проблемы
- приоритет
- статус
```

Если заявок нет, программа должна вывести сообщение:

`Заявки отсутствуют.`

### 4. Поиск заявок

Программа должна позволять искать заявки по разным параметрам.

Необходимо реализовать поиск:
```txt
- по ФИО сотрудника
- по отделу
- по статусу
- по инвентарному номеру оборудования
```

При выборе пункта `Найти заявку` пользователь должен выбрать критерий поиска, затем ввести значение для поиска.

Пример меню поиска:
```txt
1. Поиск по сотруднику
2. Поиск по отделу
3. Поиск по статусу
4. Поиск по инвентарному номеру
```
Если подходящие заявки найдены, программа выводит их список.
Если заявок не найдено, программа выводит сообщение:

`Заявки не найдены.`

### 5. Изменение статуса заявки

Программа должна позволять изменить статус существующей заявки.

Для изменения статуса пользователь вводит номер заявки, после чего выбирает новый статус:
```txt
1. новая
2. в работе
3. выполнена
```

Если заявка с указанным номером существует, программа изменяет ее статус и выводит сообщение об успешном изменении.

Если заявка не найдена, программа выводит сообщение:

`Заявка с указанным номером не найдена.`

### 6. Удаление заявки

Программа должна позволять удалить заявку.

Для удаления пользователь вводит номер заявки.

Если заявка с таким номером существует, программа удаляет ее и выводит сообщение:

`Заявка успешно удалена.`

Если заявка не найдена, программа выводит сообщение:

`Заявка с указанным номером не найдена.`

### 7. Статистика

Программа должна выводить краткую статистику по заявкам:

- общее количество заявок
- количество новых заявок
- количество заявок в работе
- количество выполненных заявок
- количество заявок с высоким приоритетом

Пример вывода:
```txt
=== Статистика ===

Всего заявок: 10
Новые: 3
В работе: 4
Выполнены: 3
С высоким приоритетом: 2
```

### 8. Сохранение данных

Программа должна сохранять данные о заявках в файл.

Формат файла можно выбрать самостоятельно:

- txt
- csv
- json

После сохранения программа должна вывести сообщение:

`Данные успешно сохранены.`

### 9. Загрузка данных

При запуске программы необходимо загружать ранее сохраненные заявки из файла, если такой файл существует.

Если файл отсутствует, программа начинает работу с пустым списком заявок.

### 10. Выход из программы

При выборе пункта `0. Выход` программа должна завершить работу.

Перед выходом желательно выполнить автоматическое сохранение данных в файл.

Требования к обработке ошибок:

1. Программа не должна завершаться аварийно при некорректном вводе пользователя.
2. Если пользователь вводит неправильный пункт меню, необходимо вывести сообщение:

`Некорректный выбор. Попробуйте снова.`

3. Если пользователь вводит несуществующий номер заявки, необходимо вывести сообщение:

`Заявка с указанным номером не найдена.`

4. Если пользователь вводит недопустимый статус или приоритет, программа должна попросить повторить ввод.

Требования к структуре программы:

В программе рекомендуется использовать:

- класс или структуру для хранения данных заявки
- список или коллекцию для хранения всех заявок
- отдельные функции или методы для:
  - вывода меню
  - добавления заявки
  - просмотра заявок
  - поиска заявок
  - изменения статуса
  - удаления заявки
  - вывода статистики
  - сохранения данных
  - загрузки данных

Пример структуры заявки:

Заявка:
- id
- employeeName
- department
- equipmentType
- inventoryNumber
- problemDescription
- priority
- status

Дополнительные требования:

1. Номер заявки должен быть уникальным.
2. После удаления заявки номера остальных заявок не должны изменяться.
3. При добавлении новой заявки номер должен быть больше, чем у всех уже существующих заявок.
4. Поиск должен работать без учета регистра букв.
5. Данные должны сохраняться между запусками программы.

Результат выполнения работы:

В результате должно быть создано консольное приложение ServiceDesk Console, которое позволяет вести учет заявок на обслуживание оборудования в организации и сохранять информацию о заявках между запусками программы.

````python
from dataclasses import dataclass
from typing import List, Optional

@dataclass
class Request:
    id: int
    employee_name: str
    department: str
    equipment_type: str
    inventory_number: str
    problem_description: str
    priority: str
    status: str = "новая"


    def __str__(self)-> str:
        return (
            f"Номер заявки: {self.id}\n"
            f"Сотрудник: {self.employee_name}\n"
            f"Отдел: {self.department}\n"
            f"Оборудование: {self.equipment_type}\n"
            f"Инвентарный номер: {self.inventory_number}\n"
            f"Описание проблемы: {self.problem_description}\n"
            f"Приоритет: {self.priority}\n"
            f"Статус: {self.status}\n"
 )
    
    @classmethod
    def from_string(cls, line: str) -> 'Request':
        parts = line.strip().split(",")
        return cls (
            id=int(parts[0]),
            employee_name = parts[1],
            department = parts[2],
            equipment_type = parts[3],
            inventory_number = parts[4],
            problem_description = parts[5],
            priority = parts[6],
            status = parts[7]
        )
    
class FileService:
    data_file = "service_requests.txt"
    
    
    @staticmethod
    def load_data() -> tuple[List[Request], int]:
        requests = []
        next_id = 1
        try:
            with open(FileService.data_file, "r", encoding="utf-8") as f:
                lines = f.readlines()
                if lines:
                    next_id = int(lines[0].strip())
                    for line in lines[1:]:
                        if line.strip():  
                            request = Request.from_string(line)
                            requests.append(request)
        except FileNotFoundError:
            pass  
        except Exception as e:
            print(f"Ошибка загрузки данных: {e}")

        return requests, next_id
    
    @staticmethod
    def save_data(requests: List[Request], next_id: int) -> bool:
        try:
            with open(FileService.data_file, "w", encoding="utf-8") as f:
                f.write(f"{next_id}\n")
                for req in requests:
                    f.write(req.to_string() + "\n")
            return True
        except Exception as e:
            print("Ошибка сохранения")
            return False
        
class RequestService: 
    
    def __init__(self):
        self._requests, self._next_id = FileService.load_data()

    def add_request(self, employee_name: str, department: str,  equipment_type: str, inventory_number: str, problem_description: str, priority: str) -> Optional[Request]:
        valid_priorities = ["низкий", "средний", "высокий"]
        if priority not in valid_priorities:
            return None
        
        request = Request (
            id=self._next_id,
            employee_name=employee_name,
            department=department,
            equipment_type=equipment_type,
            inventory_number=inventory_number,
            problem_description=problem_description,
            priority=priority
        )

        self._request.append(request)
        self._next_id += 1
        return request
    
    def get_all_requests(self) -> List[Request]:
        return self._requests
    
    def find_by_employee(self, employee_name: str) -> List[Request]:
        return [
            req for req in self._requests
            if employee_name.lower() in req.employee_name.lower()
        ]

    def find_by_department(self, department: str) -> List[Request]:
        return [
            req for req in self._requests
            if department.lower() in req.department.lower()
        ]
    
    def find_by_status(self, status: str) -> List[Request]:
        return [
            req for req in self._requests
            if status.lower() == req.status.lower()
        ]
    
    def find_by_inventory_number(self, inventory_number: str) -> List[Request]:
        return [
            req for req in self._requests
            if inventory_number.lower() in req.inventory_number.lower()
        ]
    
    def update_status(self, request_id: int, new_status: str) -> bool:
        valid_statuses = ["новая", "в работе", "выполнена"]
        if new_status not in valid_statuses:
            return False
        
        for req in self._requests:
            if req.id == request_id:
                req.status = new_status
                return True
        return False
    
    def delete_request(self, request_id: int) -> bool:
        for i, req in enumerate(self._requests):
            if req.id == request_id:
                del self._requests[i]
                return True
        return False
    
    def get_statistics(self) -> dict:
        total = len(self._requests)
        new_count = sum(1 for req in self._requests if req.status == "новая")
        in_progress = sum(1 for req in self._requests if req.status == "в работе")
        completed = sum(1 for req in self._requests if req.status == "выполнена") 
        high_priority = sum(1 for req in self._requests if req.priority == "высокий")
        
        return {
            "total": total,
            "new": new_count,
            "in_progress": in_progress,
            "completed": completed,
            "high_priority": high_priority
        }
        
    def save_all_data(self) -> bool:
        return FileService.save_data(self._requests, self._next_id)
    
class ConsoleMenu:
    def __init__(self, request_service: RequestService):
        self.request_service = request_service

    def show_menu(self):
        print("1. Добавить заявку")
        print("2. Показать все заявки")
        print("3. Найти заявку")
        print("4. Изменить статус заявки")
        print("5. Удалить заявку")
        print("6. Показать статистику")
        print("7. Сохранить данные")
        print("8. Выход")
        
    def run(self):
        while True:
            self.show_menu()
            choice = input("Выберите пункт меню: ").strip()

            if choice == "1":
                self._add_request()
            elif choice == "2":
                self._show_all_requests()
            elif choice == "3":
                self._find_requests()
            elif choice == "4":
                self._update_status()
            elif choice == "5":
                self._delete_request()
            elif choice == "6":
                self._show_statistics()
            elif choice == "7":
                self._save_data()
            elif choice == "8":
                print("До свидания!")
                break
            else:
                print("Неверный выбор. Попробуйте снова.")
                
                
    def _print_requests(self, requests: List[Request], title: str):
        if not requests:
            print("Заявки не найдены.")
            return
        
        print(f"\n{title}")
        for req in requests:
            print(req)
            print("-" * 40)
            
    def _add_request(self):
        print("\n=== ДОБАВЛЕНИЕ НОВОЙ ЗАЯВКИ ===")
        employee_name = input("ФИО сотрудника: ").strip()
        department = input("Отдел: ").strip()
        equipment_type = input("Тип оборудования: ").strip()
        inventory_number = input("Инвентарный номер: ").strip()
        problem_description = input("Описание проблемы: ").strip()
        
        while True:
            priority = input("Приоритет (низкий/средний/высокий): ").strip().lower()
            if priority in ["низкий", "средний", "высокий"]:
                break
            print("Неверный приоритет! Допустимые значения: низкий, средний, высокий")
            
        request = self.request_service.add_request(
            employee_name, department, equipment_type, inventory_number, problem_description, priority
        )
        
        if request:
            print(f"Заявка успешно добавлена!")
        else:
            print("Ошибка при добавлении заявки!")
            
    def _show_all_requests(self):
        requests = self.request_service.get_all_requests()
        if not requests:
            print("Заявки отсутствуют.")
        else:
            self._print_requests(requests, "Все заявки")
            
    def _find_requests(self):
        print("1. Поиск по сотруднику")
        print("2. Поиск по отделу")
        print("3. Поиск по статусу")
        print("4. Поиск по инвентарному номеру")
        
        choice = input("Выберите критерии поиска: ").strip()
        
        if choice == "1":
            employee_name = input("Введите ФИО сотрудника: ").strip()
            requests = self.request_service.find_by_employee(employee_name)
            self._print_requests(requests, f"Заяыки по сотруднику '{employee_name}'")
        elif choice == "2":
            department = input("Введите отдел:").strip()
            requests = self.request_service.find_by_department(department)
            self._print_requests(requests, f"Заявки по отделу '{department}'")
        elif choice == "3":
            status = input("Введите статус (новая/в работе/выполнена): ").strip()
            requests = self.request_service.find_by_status(status)
        elif choice == "4":
            inventory_number = input("Введите инвентарный номер:").strip()
            requests = self.request_service.find_by_inventory_number(inventory_number)
            self._print_requests(requests, f"ЗАЯВКИ С ИНВЕНТАРНЫМ НОМЕРОМ '{inventory_number}'")
        else:
            print("Неверный выбор критерии поиска!")
            
    def _update_status(self):
        try:
            request_id = int(input("Введите номер заявки:"))
        except ValueError:
            print("Номер заявки должен быть числом!")
            return
        
        print("Выберите новый статус:")
        print("1. новая")
        print("2. в работе")
        print("3. выполнена")
        
        status_choice = input("Ваш выбор:").strip()
        status_map = {"1": "новая", "2": "в работе", "3": "выполнена"}
        
        new_status = status_map.get(status_choice)
        if not new_status:
            print("Неверный выбор статуса!")
            return
        
        if self.request_service.update_status(request_id, new_status):
            print(f"Статус заявки №{request_id} изменён на '{new_status}'")
        else:
            print("Заявка с указанным номером не найдена.")

    def _delete_request(self):
        try:
            request_id = int(input("Введите номер заявки для удаления: "))
        except ValueError:
            print("Номер заявки должен быть числом!")
            return

        if self.request_service.delete_request(request_id):
            print("Заявка успешно удалена.")
        else:
            print("Заявка с указанным номером не найдена.")

    def _show_statistics(self):
        stats = self.request_service.get_statistics()
        print(f"Всего заявок: {stats['total']}")
        print(f"Новые: {stats['new']}")
        print(f"В работе: {stats['in_progress']}")
        print(f"Выполнены: {stats['completed']}")
        print(f"С высоким приоритетом: {stats['high_priority']}")

    def _save_data(self):
        if self.request_service.save_all_data():
            print("Данные успешно сохранены.")
        else:
            print("Ошибка при сохранении данных.")

def main():
    request_service = RequestService()
    menu = ConsoleMenu(request_service)

    print("Добро пожаловать в ServiceDesk Console!")

    try:
        menu.run()
    finally:
        
        if request_service.save_all_data():
            print("\nДанные автоматически сохранены перед выходом.")
        else:
            print("\nОшибка автоматического сохранения!")

if __name__ == "__main__":
    main()
            
        
    
        
        
        
        

            

        
   

    

````
